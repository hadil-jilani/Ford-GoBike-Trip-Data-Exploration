## Investigation Overview

In this investigation, I want to look at the indicators of the trip that could be used to predict the average duration. The main focus was on the user type, user gender and the hour of the day the trip started.

## Dataset Overview

This data set includes information about approximately 183000 individual rides made in a bike-sharing system covering the greater San Francisco Bay area. It includes features such as the start- and end-times, start- and end-locations, duration and information about the user such as their type, gender, year of birth etc..

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
bike = pd.read_csv('201902-fordgobike-tripdata.csv')
```

```python
# drop missing values
bike.dropna(inplace=True,axis=0)
```

```python
# convert user_type, member_gender, and bike_share_for_all_trip into categorical types
cat_var_dict = {'user_type':['Subscriber','Customer'],
               'member_gender':['Male','Female','Other'],
               'bike_share_for_all_trip':['Yes','No']}

for var in cat_var_dict:
    cat_var = pd.api.types.CategoricalDtype(categories = cat_var_dict[var])
    bike[var] = bike[var].astype(cat_var)
```

```python
# convert start_time and end_time to datetime format
bike['start_time'] = pd.to_datetime(bike['start_time'])
bike['end_time'] = pd.to_datetime(bike['end_time'])
```

```python
# removing entries in which member_birth_year is before 1919
bike=bike[bike['member_birth_year']>1919]
```

```python
# create a new column: hour
bike['hour'] = bike['start_time'].dt.hour
```

```python
# group the hour, user_type and the mean duration in one table
bike_hour_type = bike.groupby(['hour','user_type']).mean(numeric_only=True)
```

```python
# group the hour, member_gender and the mean duration in one table
bike_hour_gender = bike.groupby(['hour','member_gender']).mean(numeric_only=True)
```

## Distribution of Trip Durations

Trip durations in the dataset take on a very large range of values, from about *60 sec (1 min)*, to about *84000 sec (23 hours)*. Plotted on a logarithmic scale, the distribution of trip durations takes on a unimodal shape.

```python
bin_size = 70
bins = np.arange(0,85000+bin_size,bin_size)
plt.figure(figsize=[14.70, 8.27])
plt.hist(data=bike, x='duration_sec', bins=bins)
plt.xscale('log')
plt.xticks([7e1,1e2,3e2,7e2,1e3,3e3],[70,100,300,700,1000,3000])
plt.xlabel('Trip Duration (sec)')
plt.title('Distribution of Trip Duration');
#plt.xlim((60,5000));
```

## Distribution of the User Types

User of the bike sharing system are divided into two categories: *Customer* and *Subscriber*. The relative frequency distribution shows that **subscribers** have the majority with *91%* of trips, whilst the **customers** only occupy *9%*.

```python
plt.figure(figsize=[14.70, 8.27])
counts = (bike['user_type'].value_counts()).to_list()
plt.hist(data=bike, x='user_type')
plt.xlabel('User Types')
yticks=counts
ylabels=[round((counts[0]/(sum(counts))),2),
        round((counts[1]/(sum(counts))),2)]
plt.title('Distribution of User Types')
plt.yticks(yticks,ylabels)
plt.ylabel('Relative frequency');
```

## Distribution of Member Genders
Genders of users present in this dataset are *male*, *female* and *others*. The relative frequency distribution shows that **male** users have the majority with *75%* of trips, whilst the **female** only occupy *23%* and **other genders** *2%*.

```python
plt.figure(figsize=[14.70, 8.27])
counts_gen = (bike['member_gender'].value_counts()).to_list()
freq_gen = list(map(lambda x: round(x/(sum(counts_gen)),2),counts_gen))
plt.hist(data=bike, x='member_gender')
plt.xlabel('Member Genders')
yticks=counts_gen
ylabels=freq_gen
plt.yticks(yticks,ylabels)
plt.title('Distribution of Member Genders')
plt.ylabel('Relative frequency');
```

## Duration by Hour and User Type
We can see that when it comes to average duration, **customers** tend to have higher average durations than the **subscribers**. There is a particular peak between *1 am* and *4 am* fo both plots. This may indicate that the average duration will rise in this time of the day *independently*of the user type

```python
sb.set(rc = {'figure.figsize':(14.70,8.27)})
sb.lineplot(data=bike_hour_type, x="hour", y="duration_sec",hue='user_type')
xticks=np.arange(0,23+2,2)
plt.xticks(xticks,xticks)
plt.ylabel('Average Duration (sec)')
plt.xlabel('Hour of the day')
plt.title('Duration by Hour and User Type');
```

## Duration by Hour and Member Genders
Reproducing the same plot with gender instead of user type shows that the the particular peak is still there, with similar behaviour on the rest of the day (except for the other genders as there are local peaks after the bigger one). This only confirms that hour is a strong predictor of the direction of variation of the average duration plot. The values it will take can be predicted by the gender and the type of the users.

```python
sb.set(rc = {'figure.figsize':(14.70,8.27)})
sb.lineplot(data=bike_hour_gender, x="hour", y="duration_sec",
            hue='member_gender')
xticks=np.arange(0,23+2,2)
plt.xticks(xticks,xticks)
plt.ylabel('Average Duration (sec)')
plt.xlabel('Hour of the day')
plt.title('Duration by Hour and Member Gender');
```
